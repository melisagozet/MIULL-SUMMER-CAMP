# Conditions on Numpy 

```python
v = np.array([1, 2, 3, 4, 5])

#klasik yolla liste içinden 3 ten küçük elemanları yakala
ab = []

for i in v:
  if i < 3:
    ab.append(i)
```

```python
v < 3
```

```python
# Numpy ile

v[v < 3]
```

```python
print(v[v > 3])
print(v[v >= 3])
print(v[v <= 3])
print(v[v != 3])
print(v[v == 3])
```

# Mathematical Operations

```python
v = np.array([1, 2, 3, 4, 5])

print(v/5)
print(v*5/5)
print(v**5)
print(v-5)
```

```python
print(np.subtract(v, 1))  # çıkarma
print(np.add(v, 1)) # toplama
print(np.mean(v)) # ortalama
print(np.sum(v)) # liste içindeki bütün elemanların toplamı
print(np.min(v)) # listedeki en küçük eleman
print(np.max(v)) # listedeki en büyük eleman
print(np.var(v)) # varyasyonu
```

# APPLY & LAMBDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df["age2"] = df["age"]*2
df["age3"] = df["age"]*5
```

```python
df.head()
```

```python
(df["age"]/10).head()
```

```python
for col in df.columns:
  if "age" in col:
    print(col)
```

```python
#içinde age değişkeni olan bütün kolonları 10'a böl bu kod geçici olarak atadı
for col in df.columns:
  if "age" in col:
    print((df[col]/10).head())
```

```python
# kalıcı olarak atamak için

for col in df.columns:
  if "age" in col:
    df[col] = df[col]/10
df.head()
```

```python
df[["age", "age2", "age3"]].apply(lambda x: x/10).head()
```

```python
#2.yol apply lambda ile döngü yazmadan değişkenler üzerinde gezme işlemi sağladı
df.loc[:, df.columns.str.contains("age")].apply(lambda x: x/10).head()
```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x-x.mean())/ x.std()).head()
```

```python
def standart_scaler(col_name):
  return(col_name - col_name.mean()) / col_name.std()
df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
#kaydetmek için 1.yol:
df.loc[:, ["age", "age2", "age3"]] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
#kaydetmek için 2.yol:
df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

# BİRLEŞTİRME (JOIN) İŞLEMLERİ

```python
import pandas as pd
import numpy as np

m = np.random.randint(1, 30, size = (5, 3))

df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])

df2 = df1 + 99
print(m)
print(df1)
print(df2)
```

```python
pd.concat([df1, df2]) 
```

**ignore_index = True** : Birleştirilen 2.dataframe index 'i 0 dan başlatır. Bunu yok etmek için kullanılır

```python
pd.concat([df1, df2], ignore_index = True) 
```

```python
# example

df1 = pd.DataFrame({"employees": ["John", "Dennis", "Mark", "Maria"],
                    "group": ["accounting", "engineering", "engineering", "hr"]})


df2 = pd.DataFrame({"employees": ["John", "Dennis", "Mark", "Maria"],
                    "start_date": [2010, 2009, 2014, 2019]})
print(df1)
print(df2)
```

```python
pd.merge(df1, df2) #belirtmeden employees e göre birleştirdi. 
```

```python
#belirtmek istersende aşağıdaki gibi yapabilirsin.

pd.merge(df1, df2, on = "employees")
```

```python
# AMAÇ: Her çalışanın müdürünün bilgisine erişmek istiyoruz.

df3 = pd.merge(df1, df2)

df4 = pd.DataFrame({"group": ["accounting", "engineering", "hr"],
        "manager": ["Caner", "Mustafa", "Berkcan"]})
print(df4)
```

```python
pd.merge(df3, df4, on = "group")
```
